```python
#Q2.ANS
def print_exception_hierarchy(exception_class, level=0):
    indent = " " * level
    print(f"{indent}{exception_class.__name__}")

    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, level + 2)

# Start with the base Exception class
print("Python Exception Hierarchy:")
print_exception_hierarchy(Exception)
```

```python
#Q-3 ANS
'''1-overflowerror
   2-ZeroDivisionError
   
  -> OverflowError is raised when an arithmetic operation exceeds the limits of the data type being used, typically with integers or floating-point numbers.'''
  
  # Attempting to calculate a result that exceeds the maximum value of an integer
result = 2 ** 1000
print(result)

'''->ZeroDivisionError:

ZeroDivisionError is raised when attempting to divide a number by zero, which is mathematically undefined.'''
# Attempting to divide a number by zero
dividend = 10
divisor = 0

try:
    result = dividend / divisor
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

```python
#Q-4 ANS
'''The LookupError class is a base class for exceptions that are related to lookup operations,
such as indexing or searching for values within a collection, like lists, dictionaries, or sequences
Two common errors derived from the LookupError class are 'KeyError' and 'IndexError' '''
'''
KeyError:
KeyError is raised when you try to access a dictionary using a key that does not exist in the dictionary.
It typically occurs when you attempt to retrieve a value associated with a key that is not present in the dictionary.'''

# Attempting to access a non-existent key in a dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}
key = 'd'  # 'd' is not a key in the dictionary

try:
    value = my_dict[key]
except KeyError as e:
    print(f"KeyError: {e}")
    
'''IndexError:

IndexError is raised when you try to access an index in a sequence (e.g., list or tuple) that is outside the valid range of indices.
It occurs when you attempt to access an element at an index that is either too large or too small for the sequence.'''

# Attempting to access an element using an invalid index
my_list = [10, 20, 30, 40, 50]
index = 10  # This index is out of range

try:
    element = my_list[index]
except IndexError as e:
    print(f"IndexError: {e}")
```

```python
'''ImportError and ModuleNotFoundError are both exceptions in Python related to importing modules and packages.

(1) ImportError:
ImportError is a base class for exceptions that occur when there is an issue with importing a module.It is a broad exception that can encompass various problems related to importing, 
such as missing or invalid modules,issues in the imported module's code, or circular imports.''' 

try:
    import non_existent_module
except ImportError as e:
    print(f"ImportError: {e}")
    
'''In this example, we attempt to import a module called non_existent_module, which doesn't exist. This results in an ImportError.'''

'''(2) ModuleNotFoundError:

ModuleNotFoundError is a more specific exception introduced in Python 3.6. It is raised when the Python interpreter cannot find the module you are trying to import.
his error is a subclass of ImportError and provides a more specific and informative message about the missing module.'''

try:
    import non_existent_module
except ModuleNotFoundError as e:
    print(f"ModuleNotFoundError: {e}")
```
